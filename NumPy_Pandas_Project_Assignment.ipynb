## Question 1

Create a NumPy array using the salary values:

Input:
```
[50000,42000,65000,58000,45000]
```

Output:
```
array([50000,42000,65000,58000,45000])
```

````python
# Write your code here
import numpy as np
salary = np.array([50000,42000,65000,58000,45000])
salary
````

## Question 2

Using the NumPy array, find Total Salary, Average Salary and Maximum Salary.

Output:
```
Total Salary = 260000
Average Salary = 52000.0
Maximum Salary = 65000
```

````python
# Write your code here
salary = np.array([50000,42000,65000,58000,45000])
print("Total Salary =",salary.sum())
print("Average Salary =",salary.mean())
print("Maximum Salary =",salary.max())
````

## Question 3

Convert the NumPy array into a Pandas DataFrame.

Column Name: **Salary**

````python
# Write your code here
import pandas as pd
salary = np.array([50000,42000,65000,58000,45000])
df = pd.DataFrame(salary,columns=['Salary'])
df
````

Create another DataFrame containing **EmpID** and **Experience** and merge it with the Employee DataFrame.
